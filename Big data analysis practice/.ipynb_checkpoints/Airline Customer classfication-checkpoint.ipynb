#### 7.1 问题的提出
- 如何对客户分群，才能区分高价值客户、无价值客户，并对不同客户群体实施个性化的营销策略
- 餐饮行业如果通过对客户的消费行为来评价客户对企业的贡献度，提高对某些客户的关注度，以实现企业利润最大化
- 如何通过客户对菜品的消费明细，判断哪些菜品是招牌菜，哪些是配菜。提高餐饮的精准采购

#### 7.2 聚类分析
- 值没有给定划分类别的情况下，根据数据的相似度进行分组的方法。
- 分组原则是组内距离最小化，组间距离最大化。

##### K-means
- 最小化误差函数的基础上，将数据划分为预定的K类别。采用距离作为相似性评级指标。

算法过程：
1. 从N个样本数据中随机选K个作为初始聚类质心
2. 分别计算每个样本到各个聚类中心的距离，将对象分配到距离最近的聚类中。
3. 所有对象分配完成以后，重新计算K个聚类的质心
4. 跟前一次的K个聚类中心比较，如果发生变化，重复2，否则继续
5. 输出聚类结果

伪代码如下：
创建k个点作为初始聚类质心
- 当任意一个簇的质心发生变化：
    - 对数据集中每一个数据点：
        - 对每一个质心：
            - 计算质心与数据点的距离
        - 将数据点分配到距离最近的簇
    - 对每一个簇，计算簇中所有点的均值，并将均值作为质心

#### 7.3 模型建立
- 基于航空公司积累的大量客户会员信息和乘坐的航班记录，可以得到姓名、乘机间隔、次数、消费金额。
- 获取客户价值，采用RFM模型，客户消费时间间隔R，消费频率F，消费金额M这三个指标作为航空公司识别客户价值的指标
- 主要步骤：
    - 数据清洗处理
    - 利用预处理的数据，基于RFM模型进行客户分群，对各个客户群体进行特征分析，对客户进行分类
    - 针对不同类型的客户制定不同的营销策略

```python

#-*- coding:utf-8 -*-
'''
基于RFM模型使用k-means算法聚类航空消费行为特征数据
'''
import pandas as pd
data = pd.read_excel('sourcedata/i_nuc.xls',index_col = 'Id', sheet_name = 'Sheet2')
data.head()
outputfile = 'output/airlineCus.xls'

k = 3
iteration = 500
'''标准化处理数据'''
zscoredfile = 'output/zscoreddata.xls'
data_zs=1.0*(data-data.mean())/data.std() # 数据标准化
data_zs.to_excel(zscoredfile,index=False)
#print(data_zs.head())
'''使用kmeans聚类消费行为特征数据，导出各自类别的概率密度图'''
from sklearn.cluster import KMeans
model = KMeans(n_clusters = k, n_jobs = 4, max_iter = iteration)

model.fit(data_zs) # 原始数据导入Kmeans模型

r1 = pd.Series(model.labels_).value_counts() # 统计各个类别的数目 model.labels_是各个数据的label
r2 = pd.DataFrame(model.cluster_centers_) #找出聚类中心

r = pd.concat([r2,r1],axis = 1)
r.columns = list(data.columns)+[u'类别数目'] #设置列名
#print(r)

r=pd.concat([data,pd.Series(model.labels_, index= data.index)],axis =1 )
# 在原始数据的后面添加分类后的类别，保留源数据主键，axis= 1 表示横向合并矩阵

r.columns = list(data.columns)+[u'聚类类别']
#print(r)
r.to_excel(outputfile)
def density_plot(data):
    import matplotlib.pyplot as plt
    plt.rcParams['font.sans-serif']=['SongTi']
    plt.rcParams['axes.unicode_minus']=False
#    from matplotlib.font_manager import FontProperties
#    font = FontProperties(fname='/System/Library/Fonts/PingFang.ttc')
    p = data.plot(kind='kde',linewidth=2, subplots = True, sharex=False)
    [p[i].set_ylabel(u'density')for i in range(k)]
    plt.xlabel('Group%s'%(i+1))
    plt.legend()
    plt.show()
    return plt

pic_output=r'output/pd_'
for i in range(k):
    density_plot(data[r[u'聚类类别']==i]).savefig(u'%s%s.png'%(pic_output,i+1))
```
