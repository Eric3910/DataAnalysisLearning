```python
### 3.1 数据统计分析
'''
描述性统计分析
    数据的中心位置：均值mean，中位数median，众数mode
    均值和中位数用于定量数据，众数用于定性的数据
    均值相对于中位数来说，包含的信息量更大，但容易受异常值影响

'''
import pandas as pd
df = pd.read_excel('sourcedata/i_nuc.xls',
                  sheet_name='Sheet7')
#print(df.数分.describe())
#print(df.describe())
print(df.解几.size)
print(df.解几.max())
print(df.解几.min())
print(df.解几.sum())
print(df.解几.mean())
print(df.解几.var())
print(df.解几.std())

```

```python
import numpy as np

# numpy np.mean
print(np.mean(df['数分']))

# numpy np.average
print(np.average(df['数分']))

# pandas dataframe.mean()
print(df['数分'].mean())

#pandas df.median()
print(df.median())

# df.mode()
print(df.mode())
print(df)

```

```python
!!!### 3.2 分组分析
'''
根据分组字段将分析对象划分成不同的部分,对比分析各组之前差异性
常用统计指标：计数、求和、平均值
常用命令
df.groupby(by=['分类1','分类2',...])['被统计列'].agg({列别名1:统计函数1，列别名2:统计函数2,...})
'''
df = pd.read_excel('sourcedata/i_nuc.xls',sheet_name='Sheet7')
df.head(20)
df.shape
df.groupby('班级')['军训','英语','体育','性别'].mean()
## groupby 会将列名称直接作为分组对象，分组中，数值列会被聚合，非数值列会从结果中排除。当by不止一个分组对象(列名)时，需要使用list，例如：
df.groupby(['班级','性别'])['军训','英语','体育'].mean()

```

```python
## 当统计不止一个统计函数并用别名显示统计值的名称时，比如要计算平均值、标准差、总数等，可以使用agg()

df.groupby(by=['班级','性别'])['军训'].agg({'总分':np.sum,
                                          '均分':np.mean,
                                          '人数':np.size,
                                          '方差':np.var,
                                          '标准差':np.std,
                                          '最高分':np.max,
                                          '最低分':np.min})
```

```python
### 分布分析
'''
根据分析的目的，将数据（定量数据）进行等距或不等距的分组，研究各组分布规律的分析方法


'''
import pandas as pd
import numpy as np
from pandas import read_excel

df = pd.read_excel('sourcedata/i_nuc.xls',sheet_name="Sheet7")
```

```python
df['总分']=df.英语+df.体育+df.军训+df.数分+df.高代+df.解几
df['总分'].describe()
bins = [min(df.总分)-1,400,450,max(df.总分)+1]
print(bins)
labels=['less than 400','400-450','450+']
总分分层= pd.cut(df.总分,bins,labels = labels)
总分分层.head()
df['总分分层']=总分分层

df.groupby(by=['总分分层'])['总分'].agg({
                                        '人数':np.size,
})
```

```python
### 3.4 交叉分析
'''
交叉分析通常用于分析两个或两个以上分组变量之间的关系，以交叉表形式进行变量间关系的对比分析
一般分为定量、定量分组交叉
定量、定性分组交叉
定性、定性分组交叉

pivot_table(values,index,columns,aggfunc,fill_value)
value: 数据透视表中的值
index：数据透视表中的行
columns：数据透视表中的列
aggfunc:统计函数
fill_value NA值的统一替换
'''
from pandas import pivot_table
df.pivot_table(index=['班级','姓名'])
df.pivot_table(['军训','英语','体育','性别'],index=['班级','姓名'])
```

```python
df.pivot_table(values=['总分'],index=['总分分层'],columns=['性别'],aggfunc=[np.size,np.mean])
```

```python
### 3.5 结构分析
'''
分组和交叉分析基础上，计算各组成部分所占的比重，进而分析总体的内部特征的一种分析方法

'''
df
df_pt=df.pivot_table(values=['总分'],index=['班级'],columns=['性别'],aggfunc=[np.sum])
print(df_pt)
df_pt.sum(axis=1)
```

```python
df_pt.div(df_pt.sum(axis=1),axis=0)#按列占比

df_pt.div(df_pt.sum(axis=0),axis=1)#按行占比
```

```python
### 3.6 相关分析
'''
判断两个变量是否有线性相关，最直观的方法是散点图，看变量之间是否符合某个变化规律
当需要同时考察多个变量间的相关关系时，可以利用散点矩阵图同时绘制各变量间的散点图

为了更加准确描述变量之间的线性相关程度，通过计算相关系数来进行相关分析
二元变量的相关分析过程汇总，常用的有Pearson相关系数、Spearman秩相关系数和判定系数。
Pearson相关系数一般用于分析两个连续变量之间的关系，要求连续变量的取值服从正态分布。
不服从正态分布的变量、分类或等级变量之间的关联性可采用Spearman秩相关系数（也称等级相关系数）

|r| 取值
(0,0.3)低度相关
(0.3,0.8)中度相关
(0.8,1)高度相关
相关分析函数：
DataFrame.corr()
Series.corr(other)

'''
df['高代'].corr(df['数分'])
df.loc[:,['军训','英语','体育','高代','解几']].corr()
```
